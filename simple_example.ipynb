Simple example of Fourier Transform

```python
import numpy as np
from gwpy.timeseries import TimeSeries
import matplotlib.pyplot as plt
```

```python
# First we prepare a sine wave
t = np.linspace(0, 4 * np.pi, num=1000)
x = np.sin(t)
```

```python
plt.plot(t,x)
```

```python
type(x)
```

```python
# numpy.ndarray to gwpy timeseries
x_timeseries = TimeSeries(x, dt = 4 * np.pi / 1000)
```

```python
type(x_timeseries)
```

```python
plt.plot(x_timeseries)
```

```python
# Now we use gwpy to do Fourier Transform, get the frequency domain of this sine wave
x_timeseries_asd = x_timeseries.asd()
plt.plot(x_timeseries_asd)
```

The period of sin(t) is 2pi, the frequency of it is 1/2pi = 0.15915494309189535

```python
plt.plot(x_timeseries_asd)
plt.xlim(0,0.4)
```

You will see the peak at 1/2pi, means this equation is made by frequency of 1/2pi sine wave.equation

You can make a more complicate equation do the same thing

```python
t = np.linspace(0, 1000 * np.pi, num=10000)
# x_new = 3 * np.sin(t) + 5 * np.sin(0.02*t) + np.sin(0.1*t) + 10 * np.sin(0.5*t)
x_new = np.sin(t) + 3 * np.sin(3*t) + 5 * np.sin(7*t) + 10 * np.sin(8*t)
```

```python
plt.plot(t,x_new)
```

```python
x_timeseries_new = TimeSeries(x_new, dt = 1000 * np.pi / 10000)
```

```python
plt.plot(t,x_timeseries_new)
```

```python
x_timeseries_asd_new = x_timeseries_new.asd()
plt.plot(x_timeseries_asd_new)
```

You will get the amplitude of each frequency in time domain data
